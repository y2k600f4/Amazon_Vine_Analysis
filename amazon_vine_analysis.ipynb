```python
#star ratings as mixed data types
#non number replace with "0"

import numpy as np

def conv(val):
    if not val:
        return 0    
    try:
        return np.float64(val)
    except:        
        return np.float64(0)

vine_df = pd.read_csv('vine_table.csv',converters={'star_rating':conv})



```

```python
vine_df.dtypes
```

```python
#check length of vine df
len(vine_df.index)
```

```python
#Filter the data and create a new DataFrame or table to retrieve all the rows where the
#total_votes count is equal to or greater than 20 to pick reviews
#that are more likely to be helpful and to avoid having division by zero errors later on.

filtered_df = vine_df[vine_df['total_votes'] >=20]
filtered_df
```

```python
#check length of filtered df

len(filtered_df.index)
```

```python
#Filter the new DataFrame or table created in Step 1 and create a new DataFrame
#or table to retrieve all the rows where the number 
#of helpful_votes divided by total_votes is equal to or greater than 50%.

votes_df = filtered_df[filtered_df['helpful_votes']/filtered_df['total_votes'] >=.5]
votes_df
```

```python
#check length of votes df

len(votes_df.index)
```

```python
#Filter the DataFrame or table created above, and create a new DataFrame or table that retrieves
#all the rows where a review was written as part of the Vine program (paid), vine == 'Y'.
vine_program_yes = votes_df[votes_df['vine'] == 'Y'] 
vine_program_yes
```

```python
#group by star rating for vine program = Y

vine_program_yes.groupby('star_rating').count()
```
